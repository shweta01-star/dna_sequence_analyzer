```python
def _validate_dna_sequence(dna_sequence):
    """Helper function to validate if a sequence contains only valid DNA characters."""
    valid_chars = set('ATCGatcg')
    if not all(char in valid_chars for char in dna_sequence):
        raise ValueError("Invalid DNA sequence: sequence must contain only A, T, C, G characters.")

def calculate_gc_content(dna_sequence):
    """Calculates the GC content of a DNA sequence."""
    _validate_dna_sequence(dna_sequence)
    dna_sequence = dna_sequence.upper()
    length = len(dna_sequence)
    if length == 0:
        return 0.0
    gc_count = dna_sequence.count("G") + dna_sequence.count("C")
    return (gc_count / length) * 100

def transcribe_dna_to_rna(dna_sequence):
    """Transcribes a DNA sequence to an RNA sequence (replaces T with U)."""
    _validate_dna_sequence(dna_sequence)
    return dna_sequence.upper().replace("T", "U")

def reverse_complement(dna_sequence):
    """Generates the reverse complement of a DNA sequence."""
    _validate_dna_sequence(dna_sequence)
    complement_map = {'A': 'T', 'T': 'A', 'C': 'G', 'G': 'C',
                      'a': 't', 't': 'a', 'c': 'g', 'g': 'c'}

    # Generate complement sequence
    complement_seq = "".join([complement_map.get(base, base) for base in dna_sequence])

    # Reverse the complemented sequence
    return complement_seq[::-1]

# --- Example Usage ---
print("--- Mini DNA Analysis Project ---")

sample_dna = "ATGCGATACGCTTGA"
print(f"Original DNA sequence: {sample_dna}")

# 1. Calculate GC Content
gc_percentage = calculate_gc_content(sample_dna)
print(f"GC Content: {gc_percentage:.2f}%")

# 2. Transcribe DNA to RNA
rna_sequence = transcribe_dna_to_rna(sample_dna)
print(f"RNA sequence (transcription): {rna_sequence}")

# 3. Get Reverse Complement
rev_comp_dna = reverse_complement(sample_dna)
print(f"Reverse Complement DNA: {rev_comp_dna}")

print("---------------------------------")
```

```python
def translate_rna_to_protein(rna_sequence):
    """Translates an RNA sequence into an amino acid sequence (protein)."""
    genetic_code = {
        'UUU': 'F', 'UUC': 'F', 'UUA': 'L', 'UUG': 'L',
        'CUU': 'L', 'CUC': 'L', 'CUA': 'L', 'CUG': 'L',
        'AUU': 'I', 'AUC': 'I', 'AUA': 'I', 'AUG': 'M',
        'GUU': 'V', 'GUC': 'V', 'GUA': 'V', 'GUG': 'V',

        'UCU': 'S', 'UCC': 'S', 'UCA': 'S', 'UCG': 'S',
        'CCU': 'P', 'CCC': 'P', 'CCA': 'P', 'CCG': 'P',
        'ACU': 'T', 'ACC': 'T', 'ACA': 'T', 'ACG': 'T',
        'GCU': 'A', 'GCC': 'A', 'GCA': 'A', 'GCG': 'A',

        'UAU': 'Y', 'UAC': 'Y', 'UAA': '*', 'UAG': '*',
        'CAU': 'H', 'CAC': 'H', 'CAA': 'Q', 'CAG': 'Q',
        'AAU': 'N', 'AAC': 'N', 'AAA': 'K', 'AAG': 'K',
        'GAU': 'D', 'GAC': 'D', 'GAA': 'E', 'GAG': 'E',

        'UGU': 'C', 'UGC': 'C', 'UGA': '*', 'UGG': 'W',
        'CGU': 'R', 'CGC': 'R', 'CGA': 'R', 'CGG': 'R',
        'AGU': 'S', 'AGC': 'S', 'AGA': 'R', 'AGG': 'R',
        'GGU': 'G', 'GGC': 'G', 'GGA': 'G', 'GGG': 'G'
    }

    protein_sequence = []
    rna_sequence = rna_sequence.upper()

    # Iterate through the RNA sequence in codons (3 bases at a time)
    for i in range(0, len(rna_sequence) - len(rna_sequence) % 3, 3):
        codon = rna_sequence[i:i+3]
        if codon in genetic_code:
            amino_acid = genetic_code[codon]
            if amino_acid == '*': # Stop codon
                break
            protein_sequence.append(amino_acid)
        else:
            protein_sequence.append('X') # Unknown or invalid codon

    return "".join(protein_sequence)

# --- Example Usage for Translation ---
# Using the RNA sequence generated previously from 'sample_dna'
# rna_sequence = "AUGCGAUACGCUUGA"

protein = translate_rna_to_protein(rna_sequence)
print(f"RNA sequence for translation: {rna_sequence}")
print(f"Translated Protein: {protein}")
```

```python
def get_complementary_dna(dna_sequence):
    """Generates the complementary strand of a DNA sequence without reversing it."""
    _validate_dna_sequence(dna_sequence)
    complement_map = {'A': 'T', 'T': 'A', 'C': 'G', 'G': 'C',
                      'a': 't', 't': 'a', 'c': 'g', 'g': 'c'}

    # Generate complement sequence
    complement_seq = "".join([complement_map.get(base, base) for base in dna_sequence])

    return complement_seq

# --- Example Usage for Complementary DNA ---
# Using the original sample_dna
# sample_dna = "ATGCGATACGCTTGA"

complement_dna = get_complementary_dna(sample_dna)
print(f"Original DNA sequence: {sample_dna}")
print(f"Complementary DNA (not reversed): {complement_dna}")
```

### New Functionality: Motif Finding and Basic Sequence Alignment

```python
def find_dna_motifs(dna_sequence, motif):
    """Finds all occurrences of a given motif within a DNA sequence and returns their starting indices."""
    dna_sequence = dna_sequence.upper()
    motif = motif.upper()
    indices = []
    for i in range(len(dna_sequence) - len(motif) + 1):
        if dna_sequence[i:i+len(motif)] == motif:
            indices.append(i)
    return indices

# --- Example Usage for Motif Finding ---
motif_to_find = "CGA"
motif_indices = find_dna_motifs(sample_dna, motif_to_find)
print(f"Original DNA sequence: {sample_dna}")
print(f"Motif to find: {motif_to_find}")
print(f"Occurrences found at indices: {motif_indices}")
```

```python
def find_common_substrings(seq1, seq2, min_length=3):
    """Finds common substrings of a minimum length between two DNA sequences."""
    seq1 = seq1.upper()
    seq2 = seq2.upper()
    common_subs = set()

    for i in range(len(seq1)):
        for j in range(len(seq2)):
            k = 0
            while (i + k < len(seq1) and j + k < len(seq2) and
                   seq1[i + k] == seq2[j + k]):
                k += 1
            if k >= min_length:
                common_subs.add(seq1[i:i + k])
    return sorted(list(common_subs))

# --- Example Usage for Basic Sequence Alignment (Common Substrings) ---
seq_a = "ATGCGTACGTAGCTGA"
seq_b = "CGTACGTACGATCA"

print(f"Sequence A: {seq_a}")
print(f"Sequence B: {seq_b}")

common_subs = find_common_substrings(seq_a, seq_b, min_length=4)
print(f"Common substrings (min length 4): {common_subs}")

seq_c = "AATTGGCC"
seq_d = "AAGGTTCC"
common_subs_short = find_common_substrings(seq_c, seq_d, min_length=2)
print(f"\nSequence C: {seq_c}")
print(f"Sequence D: {seq_d}")
print(f"Common substrings (min length 2): {common_subs_short}")
```
